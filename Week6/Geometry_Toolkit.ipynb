```python
# IMPORT STATEMENTS
# 1. Import math
import math

# FUNCTION DEFINITIONS
  # 1. Define area(radius): returns pi * r squared
def area(radius):
  r_squared = radius ** 2
  result = math.pi * r_squared
  return result




  # 2. Define circumference(radius): returns 2 * pi * r
def circumference(radius):
  result = 2 * math.pi * radius
  return  result



  # 3. Define volume(radius, height): (find the area of the circle on the bottom)
def volume(radius, height):
  base_area = area(radius)


  #    - Call area(radius) and multiply by height.

  result = base_area * height


  #    - Return the result.
  return result

# PROGRAM LOGIC
# 1. Input for radius (float)
radius = float(input("Enter radius: "))


# 2. Call and print area and circumference.
area_quantity = area(radius)
circumference_quantity = circumference(radius)

print("area_quantity:" , area_quantity)
print("circumference_quantity:" , circumference_quantity)


# 3. Input for height (float)

height= float(input("Enter height: "))
volume_quantity = volume(radius, height)

# 4. Call and print volume.

print("volume_quantity:" , volume_quantity)
```
