```python
##### IMPORT STATEMENTS #####



###### FUNCTION DEFINITIONS #####

#Create a function called is_palindrome
def is_palindrome(word):

    #Make the word lowercase so the check is case insensitive
    word = word.lower()

    #Compare the lowercase word to the reversed word
    reversed_word = word[::-1]


    #Use string slicing to handle the reversal of the word
    #Must return true if the word is a palindrome
    #Must return false if not
    return word == reversed_word




##### PROGRAM LOGIC #####

#Prompt the user for a word and save word
user_input = input("Enter a word: ")

#Call is_palindrom using the users input
if is_palindrome(user_input):
    print(f"{user_input} is a palindrome")



#Print a message indicating the result as Trur or False
else:
    print(f"{user_input} is not a palindrome")



```
