```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define verify_pin():
def verify_pin():
  while True:
    pin = input("Enter your PIN: ")


#    - Create a loop that asks for a PIN until it equals "1234"
    if pin == "1234":
      print("PIN Correct")
      return  True

    else:
      print("PIN Incorrect")
      continue


# 2. Define withdraw(balance, amount):
def withdraw(balance, amount):



  #    - Check if enough money exists
  if amount <= balance:
    print("Withdrawal Granted")



  #    - Return the new (or old) balance
    new_balance = balance - amount
    return new_balance

  else:
    print("Insufficient Funds")
    return balance

# PROGRAM LOGIC
# 1. Set current_balance to 500
current_balance = 500

# 2. Call verify_pin()
verify_pin()


# 3. Get withdrawal amount from user input
amount = float(input("Enter amount to withdraw: "))


# 4. Update current_balance by calling withdraw()
current_balance = withdraw(current_balance, amount)


# 5. Print the final balance
print("Current Balance: $" + str(current_balance))


```
